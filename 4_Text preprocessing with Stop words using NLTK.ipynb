```python
from nltk.corpus import stopwords
```

```python
from nltk import PorterStemmer
por_stem=PorterStemmer()
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

## Tonkenization

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
sentences
```

## Applying stopwords 

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[por_stem.stem(word) for word in words if word not in set(stopwords.words("english"))]
    sentences[i]=' '.join(words)
```

## applying snowballstemmer

```python
from nltk.stem import SnowballStemmer
snowballstem=SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstem.stem(word) for word in words if word not in set(stopwords.words("english"))]
    sentences[i]=' '.join(words)
```

```python
sentences
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words("english"))]
    sentences[i]=' '.join(words)
```
