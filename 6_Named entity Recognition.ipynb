## Named entity Recognition

```python
sentences="""Incidents caused by human intervention, or man-made hazards, arise from negligence, error, or intentional acts, resulting in significant harm to life, property, and the environment. Key examples include industrial disasters like the 1984 Bhopal gas tragedy, transport accidents, environmental pollution, and structural failures. """
```

```python
import nltk
```

```python
sentences=nltk.sent_tokenize(sentences)
```

```python
sentences
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
```

```python
words
```

```python
tagged_elements=nltk.pos_tag(words)
```

```python
tagged_elements
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.ne_chunk(tagged_elements).draw()
```
