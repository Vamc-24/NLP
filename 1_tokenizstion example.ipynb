## Text preprocessing

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
corpus="""Hello my name is sai vamsi, and i'm studying in 3rd year aids! At annamacharya institute of technology.
My aim is to secure a good role in product based company with decent package. """
```

```python
## this is my corpus 
print(corpus)
```

## Tokenization 
## converting corpus into Documents

```python

Documents=sent_tokenize(corpus)
```

```python
type(Documents)
```

```python
for doc in Documents:
    print(doc)
```

## Tokenization 
## converting corpus into vocabulary

```python
Vocabulary=word_tokenize(corpus)
```

```python
for word in Vocabulary:
    print(word)
```

## converting documents into vocabulary

```python
for sentence in Documents:
    print(word_tokenize(sentence))
```

## Implement wordpunct_tokenize

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```
